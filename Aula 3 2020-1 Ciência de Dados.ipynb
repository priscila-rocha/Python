```python
import pandas as pd
import statsmodels.api as sm
import os

os.chdir("D:\Dropbox\Fund Prog e Estatistica\db")
arquivo = pd.read_csv("HBAT.csv")

nomes = ["x" + str(i) for i in range(9,19)]
print(nomes)

y = pd.DataFrame(arquivo["x19"])

for i in nomes:
    X = pd.DataFrame(arquivo[i])
    X = sm.add_constant(X)
    model = sm.OLS(y,X).fit()
    print(model.summary())
```
